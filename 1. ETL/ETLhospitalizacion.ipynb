Eliminando los valores NaN

```python
# Eliminamos todas las filas que contienen valores NaN, ya que, en conjunto solo es el 4% de toda la 
# informacion de la BBDD. Ademas que son valores no reemplazables (por un promedio, vecino mas cercano, etc)
# por ser de pruebas unicas hacia un pasciente. 

# Vemos con .info() como queda la Tabla, limpia de valores nulos, 20 cols y 541 filas
df = df.dropna()
df.info()
```

Analizando valores atípicos

```python
# analizando las filas de las edades 143 y 151 años
#vemos que solo son 2 columnas.
filas_a_eliminar = df[(df.EDAD == 143) | (df.EDAD == 151)]
filas_a_eliminar
```

Eliminando las 2 filas de valores de edad atípico

```python
#eliminamos las filas de valores 143 y 151 años, ya que en el entrenamiento nos perjudicaría por tener valores 
# no pegados a la realidad.
df = df.drop(filas_a_eliminar.index)
df.info()
```

Analizando si existen valores duplicados

```python
#analizamos si exisisten filas duplicadas, y tenemos 213 filas que se repiten
df[df.duplicated(keep=False)]
```

Eliminando valores que son duplicados

```python
# aquí eliminamos filas que esta repetidas, disminuyendo el total de datos a "". 
# si dejamos esos valores tendríamos mas datos pero nos surgiria un sesgo en la información y para nuestro 
# entrenamiento sería algo malo.

# con .info() nos damos cuenta que de 570 filas en original ahora solo tenemos 431, 24% eliminado
df = df.drop_duplicates(subset=df.columns, keep='first') # subset = que colunms analiza para encontrar duplicados
df.info() # keep = que fila dejamos si encontramos duplicados 
```
